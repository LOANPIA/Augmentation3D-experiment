```python
import os
import shutil

# # Thư mục chứa các file ground truths
# ground_truths_dir = './DulieuDr/RegisteredMask'
# # Thư mục chứa các file ảnh
# images_dir = './DulieuDr/Registered'

# Thư mục chứa các file ground truths
ground_truths_dir = './10/Labels-20240424T143505Z-001/Labels'
# Thư mục chứa các file ảnh
images_dir = './10/Images-Preprocessing/Case1-Raw'

# # Thư mục chứa các file ground truths
# ground_truths_dir = './DulieuDr/ground_truths'
# # Thư mục chứa các file ảnh
# images_dir = './DulieuDr/Processed'
# Thư mục chứa dữ liệu huấn luyện và kiểm tra
labelsTr_dir = './Finalize10/TrainLabel'
labelsVl_dir = './Finalize10/ValidLabel'
imagesTr_dir = './Finalize10/TrainData'
imagesVl_dir = './Finalize10/ValidData'

# Hàm để kiểm tra phần mở rộng của tệp
def has_extension(filename, extension):
    return filename.endswith(extension)

# Lấy danh sách tệp tin từ thư mục ground truths và images
ground_truths_files = [file for file in os.listdir(ground_truths_dir) if has_extension(file, '.nii.gz')]
images_files = [file for file in os.listdir(images_dir) if has_extension(file, '.nii.gz')]

# Sắp xếp danh sách theo thứ tự cố định (ví dụ: sắp xếp theo tên tệp tin)
ground_truths_files.sort()
images_files.sort()

# Tính toán số lượng tệp tin cho mỗi tập
num_train_files = int(len(ground_truths_files) * 0.8)
num_test_files = len(ground_truths_files) - num_train_files

# Chia danh sách tệp tin thành tập huấn luyện và tập kiểm tra
train_ground_truths_files = ground_truths_files[:num_train_files]
test_ground_truths_files = ground_truths_files[num_train_files:]
train_images_files = images_files[:num_train_files]
test_images_files = images_files[num_train_files:]

# Tạo các thư mục nếu chưa tồn tại
os.makedirs(labelsTr_dir, exist_ok=True)
os.makedirs(labelsVl_dir, exist_ok=True)
os.makedirs(imagesTr_dir, exist_ok=True)
os.makedirs(imagesVl_dir, exist_ok=True)

# Sao chép các tệp tin vào các thư mục tương ứng
for file in train_ground_truths_files:
    shutil.copy(os.path.join(ground_truths_dir, file), os.path.join(labelsTr_dir, file))
for file in test_ground_truths_files:
    shutil.copy(os.path.join(ground_truths_dir, file), os.path.join(labelsVl_dir, file))
for file in train_images_files:
    shutil.copy(os.path.join(images_dir, file), os.path.join(imagesTr_dir, file))
for file in test_images_files:
    shutil.copy(os.path.join(images_dir, file), os.path.join(imagesVl_dir, file))
```
